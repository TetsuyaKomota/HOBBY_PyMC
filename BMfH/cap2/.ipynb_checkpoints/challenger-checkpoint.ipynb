```python
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.mstats import mquantiles
```

```python
# with open して while 回すとかしなくても
# numpy.genfromtxt とかいう便利なメソッドがあったらしい
data = np.genfromtxt("tmp/challenger_data.csv", skip_header=1, usecols=[1, 2], missing_values="NA", delimiter=",")

# nan (missing_value で指定した値) を取り除く
data = data[np.isnan(data[:, 1])==False]
print(data)

plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
plt.show()
```

```python
# プロットを見ると，「気温が低いほど破損が起こりやすい」
# という傾向がありそう．MCMC で分析してみよう
#
# 破損確率が気温に対してロジスティック関数になると仮定
# ロジスティック関数は β>0 だと x=0 で 1 になる
# この定義は試してみるだけのもので別にこの先使わない
def logistic(x, alpha, beta):
    return 1/(1+np.exp(np.dot(beta, x) + alpha))

x = np.linspace(-4, 4, 100)
plt.plot(x, logistic(x, 1,  1), color="red")
plt.plot(x, logistic(x, 3, -2), color="blue")
plt.plot(x, logistic(x, -5, 7), color="green")
plt.plot(x, logistic(x, 3,  1), color="yellow")


plt.show()
```

```python
# 破損確率をロジスティック関数にする場合，
# 確率変数 α, β を親変数として持つ
# α, β の分布は正規分布にする
alpha = pm.Normal("alpha", 0, 0.001, value=0)
beta  = pm.Normal("beta" , 0, 0.001, value=0)

@pm.deterministic
def p(t=data[:, 0], alpha=alpha, beta=beta):
    return 1/(1 + np.exp(t*beta + alpha))

print(alpha.value)
print(beta.value)
print(p.value)

plt.plot(range(len(p.value)), p.value)
plt.show()

# α, β の初期値は 0 なので，
# 当然 すべての気温で 0.5 になる
```

```python
# 破損結果の観測値
obs = pm.Bernoulli("obs", p, value=data[:, 1], observed=True)

# MCMC
model = pm.Model([obs, alpha, beta])
mcmc  = pm.MCMC(model)
mcmc.sample(120000, 100000, 2)
```

```python
plt.subplot(2, 1, 1)
plt.hist(mcmc.trace("alpha")[:], bins=35, label="alpha")
plt.legend()
plt.subplot(2, 1, 2)
plt.hist(mcmc.trace("beta")[:], bins=35, label="beta")
plt.legend()
plt.show()

# 結果としては，β が気温の係数なので，これが非ゼロなら
# 破損確率は気温に依存する．特に β>0 なら負の相関があり，
# 気温が低いほど破損しやすいという仮定を裏付ける
#
# データ件数は 23件と少ないため，不確定性は大きい
```

```python
# MCMC のサンプリング結果の平均を求める事で，
# 分布の期待平均を求めることが出来る

t = np.linspace(data[:, 0].min()-5, data[:, 0].max()+5, 50)[:, None]

# あ，使った
p_t = logistic(t.T, mcmc.trace("alpha")[:, None], mcmc.trace("beta")[:, None])
mean_prob_t = p_t.mean(axis=0)

plt.plot(t, mean_prob_t)
plt.plot(t, p_t[0, :], ls="--")
plt.plot(t, p_t[-2, :], ls="--")
plt.scatter(data[:, 0], data[:, 1], alpha=0.5)

plt.show()

# 破線は「ありえたかもしれない世界」の例
# 実践はそんな世界たちの平均
```

```python
# 95% 信頼区間を求めてみよう
qs = mquantiles(p_t, [0.025, 0.975], axis=0)

plt.fill_between(t[:, 0], *qs, alpha=0.5)
plt.plot(t[:, 0], qs[0], label="95% CI", alpha=0.5)
plt.plot(t, mean_prob_t, ls="--")
plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
plt.xlim(t.min(), t.max())
plt.ylim(-0.02, 1.02)

plt.show()

# 期待平均に対して信頼区間を求める事で，
# 信頼区間の大きい領域(=不確定性の大きい領域)を
# 見つけることが出来る
# その領域のデータに集中して収集することで，
# より正確な推定ができるようになる
# ->「次にどのデータがほしいか」が分かるようになる
#
# ただ，scipy の mquantiles の働きがよくわかって
# ないので，ちゃんと調べてから使う
```

```python
# 破損確率のモデルにロジスティック関数を仮定したが，
# これが適切かどうか調べるにはどうすればいいか？
# モデルが適切である 
# <=> そのモデルから生成されるデータとして自然である
# ということなので，実際にそのモデルからデータを生成
# させてみて，見比べると良い
#
# データを生成させるには，観測 obs の値を固定
# しなければよい
# 今回観測は以下の形だったので，
# obs = pm.Bernoulli("obs", p, value=data[:, 1], observed=True)
# シミュレーションはこのようにすればOK
simulated = pm.Bernoulli("simulated", p)

# MCMC
mcmc  = pm.MCMC([obs, alpha, beta, simulated])
mcmc.sample(10000)
```

```python
for i in range(4):
    plt.subplot(4, 1, i+1)
    plt.scatter(data[:, 0], mcmc.trace("simulated")[:].astype("int")[1000 * i, :], alpha=0.5)
plt.show()

# シミュレーション結果が本物そっくりで，
# なおかつそれぞれ似ているので，モデルは適切だと判断する
```

```python
# 演習問題2
# α に対する β の分布がどうなっている?

plt.scatter(mcmc.trace("alpha")[:], mcmc.trace("beta")[:])
plt.show()

# なんかくっそキレイな負の相関がある．なにこれ？
# ・α が大きいほど β が小さい
# ・αが大きいほど変化部分が左による
# ・βが小さいほど変化が緩やか
# あー，なるほど，「何としても『全体的に右下がり』」を
# 実現するためには，変化部分が左によるほど緩やかにしないと
# いけないんだね
# 右下がり <=> 事故率が気温に負の相関 ということなので
# ここからも気温依存性が見えてくるんだね
```
